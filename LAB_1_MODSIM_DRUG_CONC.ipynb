```
import pandas as p
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```

k=0.1386
t=np.linspace(0,24,24*60 +1)
z=[];a=[];b=[];
z.append(6)
a.append(6)
b.append(6)
for i in range (0,24*60):
  z.append(z[i] - k*z[i]/60)
  a.append(a[i] - 0.16*a[i]/60)
  b.append(b[i] - 0.10*b[i]/60)
print(t)
plt.plot(t,b,label="k=0.10")
plt.plot(t,z,label="k=0.1386")
plt.plot(t,a,label="k=0.16")
plt.xlabel('Time (min)')
plt.ylabel('Concentration (unit)')
plt.legend()
```
